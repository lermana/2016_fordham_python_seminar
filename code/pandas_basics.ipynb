# Working with one stock's data

#### Notes:

The read_csv() function will read a csv file into a DataFrame. If the csv file has a header row, there is no
need to pass a value for the header argument. In this case, the file has no header row, so we need to tell the function not to take the first set of values as the column headers. 

```python
aapl = pandas.read_csv('../stock_csvs/table_aapl.csv', header=None)
```

#### Notes:
    
The head() method shows us by default the first five rows of the DataFrame across all columns. Passing an integer to head() will show that number of rows.

```python
aapl.head()
```

#### Notes:

We can set column names by passing a list of names to the columns attribute. The number of names here must be equal to the number of columns in the DataFrame.

```python
aapl.columns = ['date', 'time', 'open', 'high', 'low', 'close', 'volume']
```

```python
aapl.head()
```

#### Notes: 

a DataFrame's index is used for a number of things, including:

- retrieving particular rows of a DataFrame
- joining DataFrames together

By default, a DataFrame will be given an incremental integer index that starts at 0. You can change the index easily to take the vaules of one of the DataFrame's columns with the set_index() method. User can select whether to remove the column, after passing it as an index, via the drop argument. 

A lot of DataFrame methods will return a new DataFrame. User can select for this to happen to the existing DataFrame instead by passing True to the inplace argument when it is avalable. Note that this does not actually guarantee a more efficient process from a memory standpoint. It is more of a matter of preference or code style. 

```python
aapl.set_index('date', drop=True, inplace=True)
```

```python
aapl.head()
```

```python
aapl.time.sum()
```

#### Notes:
    
You can drop columns from a DataFrame using the drop() method. In this case, our time column is not useful, so we will drop it. Passisng axis=1 is necessary so that the function knows to search columns labels, and not row labels, for the value 'time'. (drop() can also be used to remove rows, and the function sets a default value of 0 for axis)

```python
aapl.drop('time', axis=1, inplace=True)
```

```python
aapl.head()
```

#### Notes:
    
DataFrames include a number of methods for describing their data. 

The function max() returns the maximum for each column in the DataFrame. 

The function describe() returns a count of non-null rows, as well as the mean, standard deviation, min, max and quartiles, for each column of the DataFrame. 

```python
aapl.max()
```

```python
aapl.describe()
```

#### Notes:
    
Many DataFrame frame methods also work for the Series, or DataFrame column.

```python
aapl['close'].max()
```

```python
aapl['close'].describe()
```

#### Notes:

The simplest way to call a DataFrame's column is, if we have an exmaple DataFrame named df, with a column 'col':

df['col']

You can also use this syntax to create new columns. Additionally, you can access more than one column at a time by using double brackets:

df[['col1', 'col2']]

```python
open_close_avg = aapl[['open', 'close']].mean(axis=1)
```

```python
open_close_avg.head()
```

#### Notes:
    
Let's add a daily return column to our aapl DataFrame. 

First we'll need the difference from one day to the next, which we'll calculate using the shift() method. 

Next we'll divide this difference by the previous day's closing price. 

```python
aapl['delta'] = aapl['close'] - aapl['close'].shift(1)
```

```python
aapl['return'] = aapl['delta'] / aapl['close']
```

```python
aapl.head()
```

```python
aapl.drop('delta', axis=1, inplace=True)
```

```python
aapl.head()
```

#### Notes:
    
Writing a DataFrame to a .csv file is as simple as calling the to_csv() method and passing a file path. 

```python
aapl.to_csv('../output/aapl_new.csv')
```
